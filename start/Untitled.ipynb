 
 # Data Visualization
 

```python
#preapre some data
y = [ 6, 7 ,2, 5, 3]
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(y)
plt.show()
```

```python
# create a figure
plt.figure(figsize = (10 , 5))

# add need to make the figure eadable ---very important form stats prespective!

plt.title("first figure example plot ")
plt.xlabel(" x vals (S)")
plt.ylabel("y vals (m)")
plt.grid()

plt.show()
```

```python
## plotting weights 
# create figures

plt.figure(figsize=(20 ,5))

# add a line renfers with legends and line thinkness
x  = range(da["BMXWT"].size)
print(x)
y  = da["BMXWT"]

plt.scatter(x , y , marker='x' , s= 30)

# also need to make the figure readable --very important from stat prospective

plt.title("wights of individuals")
plt.xlabel("ID of individuals")
plt.ylabel("weights (kg)")

plt.show()
```
